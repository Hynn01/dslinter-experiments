# Customising Figure Asthetics in Seaborn

## Loading libraries

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Sin-wave function

```python
def sinplot(flip=1, start = 0, step = 14, end = 100, r_start=1, r_end=7):
    x = np.linspace(start,step,end)
    for i in range(r_start,r_end):
        plt.plot(x, np.sin(x+i*.5)*(7-i)* flip)
```

```python
sinplot(1)
```

```python
sinplot(5,5)
```

```python
sinplot(3, 5, 1)
```

```python
sns.set_theme()
sinplot()
```

## Seaborn Figure Style

Set style to **darkgrid**, there are five seaborn themes:
1. darkgrid
2. whitegrid
3. dark
4. white
5. ticks

```python
sns.set_style('dark')
```

Load the data randomly

```python
data = np.random.normal(size=(20,6)) + np.arange(6)/2
```

Box Plot with Seaborn

```python
sns.boxplot(data=data)
```

```python
sns.set_style('darkgrid')
```

```python
sinplot()
```

## Removing Axes Spines

```python
sinplot(4,5)
sns.despine()
```

```python
fig, axis = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True)
```
