```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style='darkgrid')
```

```python
sns.scatterplot(data=df, x='G3Dmark', y='TDP')
# It's a mess and I don't know what it is.
```

```python
sns.relplot(data=df, kind='scatter', x='G3Dmark', y='TDP', col='category', col_order=['Mobile', 'Desktop', 'Workstation'])
# Separate scatter plots by category. This is easier to understand.
```

```python
# We further divide the scatterplot by generation, grouping them by three years.
def convert_year(y):
    a = 0
    if y <= 2011:
        a = 2011
    elif y <= 2014:
        a = 2014
    elif y <= 2017:
        a = 2017
    elif y <= 2020:
        a = 2020
    else:
        a = 2022
    return a
df['generation'] = df.testDate.apply(lambda x: convert_year(x))
```

```python
df.head(10)
```

```python
sns.relplot(data=df, kind='scatter', x='G3Dmark', y='TDP',
            col='category',   col_order=['Mobile', 'Desktop', 'Workstation', 'Mobile, Workstation', 'Desktop, Mobile', 'Unknown'],
            row='generation', row_order=[2022, 2020, 2017, 2014, 2011])
# The improvement in GPU performance can now be seen in the scatter plots.
```

```python
sns.relplot(data=df, kind='scatter', x='G3Dmark', y='TDP',
            col='category',   col_order=['Mobile', 'Desktop', 'Workstation', 'Mobile, Workstation', 'Desktop, Mobile', 'Unknown'],
            hue='generation')
# Another view of the improvement in GPU performance.
```
