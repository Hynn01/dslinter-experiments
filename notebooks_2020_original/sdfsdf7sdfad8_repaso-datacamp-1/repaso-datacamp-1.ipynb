# list slicing🦢

*inclusive -> [start:end] <- exclusive*

```python
y = [1, 2, 3, 4, 99]

y[0:1]
```

```python
y = [1, 2, 3, 4, 99]

y[0:2]
```

```python
y = [1, 2, 3, 4, 99]

y[1:-1]
```

```python
y = [1, 2, 3, 4, 99]

y[-3:-1]
```

```python
y = [1, 2, 3, 4, 99]

y[:]
```

```python
y = [1, 2, 3, 4, 99]

y[:3]
```

```python
y = [1, 2, 3, 4, 99]

y[3:]
```
